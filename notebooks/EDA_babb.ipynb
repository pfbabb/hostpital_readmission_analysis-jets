```python
patient = admis[admis['Patient ID']=='DS1|10451'].sort_values('Encounter End Date').drop_duplicates(subset=['Patient ID','Encounter Start Date'], keep='last')
patient
```

```python
trigger=patient.iloc[:,8]<=(patient.iloc[0,9] + pd.DateOffset(days=30))
trigger
```

```python
readmit = 0
skipdate = pd.to_datetime(0)

for rnum in range(0,len(patient)):
    if skipdate < patient.iloc[rnum,8] :
        print('rnum:' + str(rnum))
        trigger = patient.iloc[rnum+1:,8]<=(patient.iloc[rnum,9] + pd.DateOffset(days=30))
        print(int(trigger.sum() > 0))
        readmit = readmit + int(trigger.sum() > 0)
        skipdate = (patient.iloc[rnum,9] + pd.DateOffset(days=30))
    print(str(rnum)+'in:'+str(patient.iloc[rnum,8])
          +', out+30:'+ str(patient.iloc[rnum,9]+ pd.DateOffset(days=30)))
```
